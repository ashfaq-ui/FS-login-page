```python
import pandas as pd
from sklearn import preprocessing

# 1. Loading the dataset
# Make sure you've uploaded Mall_Customers.csv to your Files sidebar first
data_frame = pd.read_csv('/content/Mall_Customers.csv')

# 2. Exploring the data
print("--- Initial Data ---")
print(data_frame.head())
print("\nDimensions:", data_frame.shape)
print("\nMissing Values:\n", data_frame.isnull().sum())

# 3. Renaming columns for easier access
data_frame.rename({
    'CustomerID': "ID",
    'Gender': "Sex",
    'Annual Income (k$)': "Salary",
    'Spending Score (1-100)': "Spending_Score"
}, axis=1, inplace=True)

# 4. Removing the unnecessary ID column
data_frame.drop('ID', axis=1, inplace=True)

# 5. Encoding 'Sex' to numeric values (0 and 1)
label_encoder = preprocessing.LabelEncoder()
data_frame['Sex'] = label_encoder.fit_transform(data_frame['Sex'])

# 6. Creating a satisfaction function
def satisfaction(value):
    if value > 70:
        return "Satisfied"
    else:
        return "Unsatisfied"

# 7. Applying the function to a new column
data_frame['Customer_Satisfaction'] = data_frame['Spending_Score'].apply(satisfaction)

# 8. Final check and saving the file
print("\n--- Processed Data ---")
print(data_frame.head())

data_frame.to_csv('/content/Prepared_Mall_Customers.csv', index=False)
print("\nSuccess: 'Prepared_Mall_Customers.csv' created in the Files sidebar.")
```
